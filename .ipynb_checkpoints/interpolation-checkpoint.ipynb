# Load Two data

```python
import os
import imageio
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import torch
import torch.nn as nn

def normalize(input_x, new_max=255, new_min=0):
    return ((new_max - new_min) * ((input_x - input_x.min()) / (input_x.max() - input_x.min()))) + new_min

data_dir = './results/gaugan_cycle/test_latest/images/'
save_dir = './results/gaugan_interpolation/sample3'
os.makedirs(save_dir, exist_ok=True)

image_a = '00000002_real_A.png'
image_b = '00000003_(2)_real_A.png'

# Just Mix
image_a_arr = imageio.imread(os.path.join(data_dir, image_a))
image_b_arr = imageio.imread(os.path.join(data_dir, image_b))
image_a_b_mixed_arr = normalize(image_a_arr + image_b_arr)

plt.hist(image_a_b_mixed_arr.flatten())
plt.show()

imageio.imwrite(os.path.join(save_dir, image_a), image_a_arr)
imageio.imwrite(os.path.join(save_dir, image_b), image_b_arr)
```

# Extract Latent Vector

```python
# load Encoder, Decoder(Generator)

from options.test_options import TestOptions
from models import networks
import cv2

def get_activation(name):
    def hook(model, input, output):
        activation[name] = output.detach()
    return hook

def get_latent_value(generator, input_x: torch.Tensor):
    activation = {}
    latent_hook_layer = generator.model.model[1].model[3].model[3].model[3].model[3].model[3].model[3]
    latent_hook_layer.model.register_forward_hook(get_activation('latent'))
    output = generator(input_x).float()
    return activation['latent']


def get_latent_value_cycle(generator, input_x: torch.Tensor):
    activation = {}
    def get_activation(name):
        def hook(model, input, output):
            activation[name] = output.detach()
        return hook
    latent_hook_layer = generator.model[18]
    latent_hook_layer.register_forward_hook(get_activation('latent'))
    output = generator(input_x).float()
    return activation['latent']


def get_skip_value(generator, input_x: torch.Tensor):
    pass


def convert_output_to_rgb(output: torch.Tensor) -> np.ndarray:
    output = np.transpose(output.detach().cpu().numpy()[0], (1, 2, 0))
    output = (255 * output).astype(np.uint8)
    return output


def change_latent(latent: torch.Tensor):
    raise NotImplementedError
    

class Identity(nn.Module):
    def __init__(self):
        super(Identity, self).__init__()
        
    def forward(self, x):
        return x

    
generator = networks.define_G(3, 3, 64, 'resnet_9blocks', 'instance')
loaded_weight = torch.load("./checkpoints/gaugan_cycle/latest_net_G_A.pth")
generator.load_state_dict(loaded_weight)

generator.eval()
with torch.no_grad():
    image_a_label = np.expand_dims(np.transpose(image_a_arr, (2, 0, 1)), 0)
    image_a_gen = generator(torch.from_numpy(image_a_label).float())
    latent_a_value = get_latent_value_cycle(generator, torch.from_numpy(image_a_label).float())
    
    image_b_label = np.expand_dims(np.transpose(image_b_arr, (2, 0, 1)), 0)
    image_b_gen = generator(torch.from_numpy(image_b_label).float())
    latent_b_value = get_latent_value_cycle(generator, torch.from_numpy(image_b_label).float())
    
    image_ab_label = np.expand_dims(np.transpose(image_a_b_mixed_arr, (2, 0, 1)), 0)
    image_ab_gen = generator(torch.from_numpy(image_ab_label).float())
    latent_ab_value = get_latent_value_cycle(generator, torch.from_numpy(image_ab_label).float())
    
    latent_a_b_mixed_value = torch.lerp(latent_a_value, latent_b_value, 0.5)
    
    print(latent_a_value.size(), latent_b_value.size(), latent_a_b_mixed_value.size())

#     plt.imshow(np.transpose(image_a_gen.numpy()[0], (1, 2, 0)), cmap='gray')
    plt.imshow(convert_output_to_rgb(image_a_gen), cmap='gray')
    imageio.imwrite(os.path.join(save_dir, f"Image_A_gen.png"), np.transpose(image_a_gen.numpy()[0], (1, 2, 0)))
    plt.colorbar()
    plt.show()
    plt.close()
    
    plt.imshow(convert_output_to_rgb(image_b_gen), cmap='gray')
    imageio.imwrite(os.path.join(save_dir, f"Image_B_gen.png"), np.transpose(image_b_gen.numpy()[0], (1, 2, 0)))
    plt.colorbar()
    plt.show()
    plt.close()
    
    plt.imshow(convert_output_to_rgb(image_ab_gen), cmap='gray')
    imageio.imwrite(os.path.join(save_dir, f"Image_AB_gen.png"), np.transpose(image_ab_gen.numpy()[0], (1, 2, 0)))
    plt.colorbar()
    plt.show()
    plt.close()
    
# Generate with Latent-Interpolation values

def get_cyclegan_decoder(G):
    for i in range(19):
        G.model[i] = Identity()
    return G

for latent_ratio in range(1, 10):
    latent_a_b_mixed_value = torch.lerp(latent_a_value, latent_b_value, latent_ratio * 0.1)

    generator_decoder = get_cyclegan_decoder(generator)
    image_ab_gen_interp = generator_decoder(latent_a_b_mixed_value)
    imageio.imwrite(os.path.join(save_dir, f"Landscape_AB_gen_interpolation_{latent_ratio}.png"), np.transpose(image_ab_gen_interp.detach().numpy()[0], (1, 2, 0)))

#     plt.imshow(convert_output_to_rgb(image_ab_gen_interp), cmap='gray')
#     plt.colorbar()
#     plt.show()
#     plt.close()
```

```python
# Load generated images

generated_image_dir = './results/gaugan_atob_pix2pix/test_latest/images/'
target_image_name = '00000007'

real_A_image = os.path.join(generated_image_dir, target_image_name+'_real_A.png')
real_B_image = os.path.join(generated_image_dir, target_image_name+'_real_B.png')
fake_B_image = os.path.join(generated_image_dir, target_image_name+'_fake_B.png')

plot(imageio.imread(real_A_image), "Real A")
plot(imageio.imread(real_B_image), "Real B")
plot(imageio.imread(fake_B_image), "Fake B")
```

```python
# Convert fake label into mathced values

def find_most_probable_id(value):
    all_ids = np.array([x.get('id') for x in COLOR_MAP.values()])
    return all_ids[np.argmin(np.abs(all_ids - value))]

def convert2id(image: np.ndarray) -> np.ndarray:
    image_converted = np.zeros_like(image).flatten()
    for pixel_idx, pixel_value in enumerate(image.flatten()):
        converted_pixel_value = find_most_probable_id(pixel_value)
        image_converted[pixel_idx] = converted_pixel_value
#         print(f"Original Value {pixel_value} -> Converted id {converted_pixel_value}")
#         break
        
    return np.reshape(image_converted, image.shape)

plot(imageio.imread(real_A_image), title='real A ')
plot(imageio.imread(fake_B_image), title='fake B (Before)')
plot(convert2id(imageio.imread(fake_B_image)), title='fake B (After)')

plt.hist(imageio.imread(fake_B_image).flatten(), bins=100)
plt.title("Fake Label Histogram (Before id matching)")
plt.show()
plt.close()

plt.hist(convert2id(imageio.imread(fake_B_image).flatten()), bins=100)
plt.title("Fake Label Histogram (After id matching)")
plt.show()
plt.close()
```
