# Convert Landscape

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import imageio
import cv2
from glob import glob

%matplotlib inline

def plot(image: np.ndarray, title: str):
    plt.imshow(image)
    plt.title(title)
    plt.colorbar()
    plt.show()
    plt.close()
```

```python
def gaugan_infer_setting(image: np.ndarray) -> np.ndarray:
    # reshape
    image = cv2.resize(image, (256, 256))
    # make label
    label = np.zeros_like(image)
    return np.concatenate([image, label], 1)


def cyclegan_infer_setting(image: np.ndarray) -> np.ndarray:
    # reshape
    image = cv2.resize(image, (256, 256))
    label = np.zeros_like(image)
    return image, label


def mix_image(image1: np.ndarray, image2: np.ndarray) -> np.ndarray:
    return image1 + image2
    

original_dir = "./datasets/landscape/original/"
save_dir = "./datasets/landscape_cycle/testA"

original_image_list = glob(os.path.join(original_dir, '*.jpg'))

sample_image = original_image_list[0]
mix_image = original_image_list[2]

plot(imageio.imread(sample_image), os.path.basename(sample_image))
# plot(gaugan_infer_setting(imageio.imread(sample_image)), os.path.basename(sample_image))

plot(imageio.imread(mix_image), os.path.basename(sample_image))
# plot(gaugan_infer_setting(imageio.imread(sample_image)), os.path.basename(sample_image))
```

```python
# Convert all

from tqdm import notebook

for image in notebook.tqdm_notebook(original_image_list):
    image_arr, label_arr = cyclegan_infer_setting(imageio.imread(image))
    dst_A = os.path.join(save_dir, os.path.basename(image))
    imageio.imwrite(dst_A, image_arr)
    
    dst_B = os.path.join(save_dir.replace("testA", "testB"), os.path.basename(image))
    imageio.imwrite(dst_B, label_arr)
#     break
```
