```python
import matplotlib.pyplot as plt
import numpy as np
import os
import imageio
from scipy import misc
from glob import glob

import cv2


def bw2rgb(image: np.ndarray) -> np.ndarray:
    return np.repeat(np.expand_dims(image, 2), 3, axis=2)


def make_new_name(current_name):
    status = os.path.dirname(gen_data).split('/')[-1].split('_')[1]
    return f'{status}_{os.path.basename(gen_data)}'


data_dir = "./datasets/horse2zebra/trainA"
data_list = glob(os.path.join(data_dir, '*.jpg'))

for data in data_list:
    data_arr = imageio.imread(data)
    plt.imshow(data_arr)
    plt.title(f'{os.path.basename(data)} - {data_arr.shape}')
    plt.show()
    plt.close()
    
    break
```

```python
save_dir = './datasets/gaugan_cycle/'

gaugan_generated_dir = "/data01/swkim/gaugan_data/gaugan_generated/2_neutral/"
gaugan_labels_dir = "/data01/swkim/gaugan_data/gaugan_labels/2_neutral/"

gen_list = sorted(glob(os.path.join(gaugan_generated_dir, '*.png')))
lbl_list = sorted(glob(os.path.join(gaugan_labels_dir, '*.png')))

num_train, num_test = 60, 19

for img_idx, (gen_data, lbl_data) in enumerate(zip(gen_list, lbl_list)):
    
    phase_save_dir = os.path.join(save_dir, 'train') if img_idx < num_train else os.path.join(save_dir, 'test')
    
    save_path_A = os.path.join(phase_save_dir+'A', make_new_name(gen_data).replace('png', 'jpg'))
    save_path_B = os.path.join(phase_save_dir+'B', make_new_name(gen_data).replace('png', 'jpg'))
    
    gen_arr = plt.imread(gen_data)
    lbl_arr = plt.imread(lbl_data)
    
    gen_arr = cv2.resize(gen_arr, (256, 256))
    lbl_arr = cv2.resize(lbl_arr, (256, 256))
    
    lbl_arr = bw2rgb(lbl_arr)
        
#     concat_arr = np.concatenate([gen_arr, lbl_arr], axis=1)

    imageio.imwrite(save_path_A, gen_arr)
    imageio.imwrite(save_path_B, lbl_arr)
    
#     break
```
